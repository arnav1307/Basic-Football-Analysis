```python
#radar charts

#importing the packages
import pandas as pd
import matplotlib.pyplot as plt
from soccerplots.radar_chart import Radar
```

```python
#read the data 
#https://fbref.com/en/comps/9/shooting/Premier-League-Stats#all_stats_shooting download the csv
df = pd.read_csv('radars.csv')
```

```python
df
```

```python
df.Player.unique()
```

```python
#removing two names in Player column
df['Player'] = df['Player'].str.split('\\',expand = True)[0]
```

```python
#choosing any two players I want basically creating a new dataset
df = df[(df['Player'] == 'Harry Kane') | (df['Player'] == 'Bruno Fernandes')].reset_index()

#dropping the columns that are not reqd
df = df.drop(['index','Rk','Nation','Pos','Squad','Age','Born','90s','Matches'],axis=1)
```

```python
df.head()
```

```python
#getting the parameters
params = list(df.columns)
params = params[1:]
params
```

```python
#adding the ranges for both the players to the list 
ranges = []
a_values = []
b_values = []

for x in params:
    a = min(df[params][x])
    a = a - (a*0.25)
    
    b = max(df[params][x])
    b = b + (b*0.25)
    
    ranges.append((a,b))

for x in range(len(df['Player'])):
    if df['Player'][x] == 'Harry Kane':
        a_values = df.iloc[x].values.tolist()
    if df['Player'][x] == 'Bruno Fernandes':
        b_values = df.iloc[x].values.tolist()
        
 #removing the name and startig the list from the first column       
a_values = a_values[1:]
b_values = b_values[1:]

values = [a_values,b_values]
```

```python
values
```

```python
#CREATION OF PLOT

#title

title = dict(
    title_name = 'Harry Kane',
    title_color = 'blue',
    subtitle_name = 'Spurs',
    subtitle_color = 'blue',
    title_name_2 = 'Bruno Fernandes',
    title_color_2 = 'red',
    subtitle_name_2 = 'Manchester United',
    subtitle_color_2 = 'red',
    title_fontsize = 18,
    subtitle_fontsize = 15
    )

endnote = '@arny \n data from FBref'

#creation of radar
radar = Radar(background_color="#121212", patch_color="#28252C", label_color="#F0FFF0",
              range_color="#F0FFF0")

fig,ax = radar.plot_radar(ranges=ranges, params=params, values=values,
                          radar_color=['blue','red'],
                          alphas=[0.75,0.6],
                          title = title,
                          endnote=endnote,compare=True
                          )
```
