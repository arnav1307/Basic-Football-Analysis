```python
#Moving Average basically gives us an idea what the average is over a period of time.Tells us the trend as well
#better rather than  plotting xG plots
#import the library 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#dataset
df = pd.read_csv('tottenham prem xg.csv')
```

```python
df
```

```python
#new column
df['match'] = range(1,39)
```

```python
df.head()
```

```python
#more columns for xgsma: expected goals simple moving average for every 3 matches
df['xgSMA'] = df['xG'].rolling(window=3).mean()
#more columns for xgasma: expected goals against simple moving average for every 3 matches
df['xgaSMA'] = df['xGA'].rolling(window=3).mean()
```

```python
df.head(10)
```

```python
#plotting the graph
plt.plot(df.match,df.xgSMA,color = 'blue')
plt.plot(df.match,df.xgaSMA,color = 'red')
```
