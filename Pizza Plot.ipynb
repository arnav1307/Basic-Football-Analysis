```python
#Pizza plot or charts: another form of the radar charts and helps us to go in more depth for a particular file 

#packages
import numpy as np
import pandas as pd

from scipy import stats
import math

from mplsoccer import PyPizza,add_image,FontManager
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('PL_defensive.csv')
```

```python
df.head()
```

```python
#removing one name by splitting the name column 
df['Player'] = df['Player'].str.split('\\',expand = True)[0]
```

```python
df.head()
```

```python
#data preprocessing 
#taking players in consideration and players more than 15 90's(regular players)
df = df.loc[(df['Pos'] == 'DF') & (df['90s']>15)]
df.head()
```

```python
df.shape
```

```python
#dropping the columns that don't have any meaning in our plot 
df = df.drop(['Rk','Nation','Pos','Squad','Age','Born'],axis=1).reset_index()
df.head()
```

```python
#creating the parameter list
params = list(df.columns)
params
```

```python
#dropping player,index for comparision
params = params[2:]
params
```

```python
#filtering the player that we want analyze
player = df.loc[df['Player']== 'Luke Shaw'].reset_index()
player = list(player.loc[0])
print(player)
```

```python
#you can choose any defender from this list 
df.Player.values
```

```python
#dont need the first three values so we will use slicing 
#Luke Shaw
print(len(player),print(len(params)))
player = player[3:]
print(len(player),print(len(params)))
```

```python
#calculating the percentile scores from the raw data 
values=[]
for x in range(len(params)):   
    values.append(math.floor(stats.percentileofscore(df[params[x]],player[x])))
```

```python
round(stats.percentileofscore(df[params[0]],player[0]))
```

```python
values
```

```python
# instantiate PyPizza class
baker = PyPizza(
    params=params,                  # list of parameters
    straight_line_color="#000000",  # color for straight lines
    straight_line_lw=1,             # linewidth for straight lines
    last_circle_lw=1,               # linewidth of last circle
    other_circle_lw=1,              # linewidth for other circles
    other_circle_ls="-."            # linestyle for other circles
)
```

```python
# plot pizza
fig, ax = baker.make_pizza(
    values,              # list of values
    figsize=(8, 8),      # adjust figsize according to your need
    param_location=110,  # where the parameters will be added
    kwargs_slices=dict(
        facecolor="red", edgecolor="#000000",
        zorder=2, linewidth=1
    ),                   # values to be used when plotting slices
    kwargs_params=dict(
        color="#000000", fontsize=12,
        va="center", alpha=.5
    ),                   # values to be used when adding parameter
    kwargs_values=dict(
        color="#000000", fontsize=12,
        zorder=3,
        bbox=dict(
            edgecolor="#000000", facecolor="grey",
            boxstyle="round,pad=0.2", lw=1
        )
    )                    # values to be used when adding parameter-values
)

# add title
fig.text(
    0.515, 0.97, "Luke Shaw - Manchester United", size=18,
    ha="center", color="#000000"
)

# add subtitle
fig.text(
    0.515, 0.942,
    "Per 90 Percentile Rank vs Premier League Defenders | 2020-21",
    size=15,
    ha="center", color="#000000"
)
```
